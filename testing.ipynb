```python
with open(base_path+"/"+sample_name, "r") as f:
    raw_str = f.read()
data = json.loads(raw_str)
```

```python
info = {
    "workload":data["name"],
    "batch_size":data["batch_size"],
    "codegen":data["target"],
    "device":data["device"],
    "total_time":data["total_time"]
}
```
